```python
import random

#A)
def generat(n,a,b,myfile):

    with open(myfile,'w') as ffile:
        for i in range(n):
            number = random.randint(a,b)
            ffile.write(ffile"{number} ")

#B)
def read(myfile):
    to_return=[]
    with open(myfile,'r') as ffile:
        to_return = ffile.read().split(' ')
    return to_return

def binary(arr,n,l,r_:
    while(l<=r):
        m=int( (l+r)//2 )
        if(arr[m] == n):
            return m
        elif(arr[m] < n):
            l=m+1
        else:
            r=m-1
    return None

#C)
def total(file1,file2,k):
    data1 = read(file1)
    data2 = read(file2)
    
    finalresut=[]

    data1.sort()
    for i2 in data2:
        i1 = k - i2 ## since  i1 + i2 = k
        search_result = binary(data1,i1,l=0,r=len(data1)-1)
        if(search_result == None):
            continue
        else:
            finalresut.append( (i1,i2) )
    return finalresut

def testing(n1,a1,b1, n2,a2,b2, k):
    generat(n1,a1,b1,str( hash('file1') ))
    generat(n2,a2,b2,str( hash('file2') ))

    data1 = read(str( hash('file1')) )
    data2 = read(str( hash('file2')) )

    print(total( str( hash('file1')), str( hash('file2')) ))

#D)
print('testing1')
testing(2000,1,10000, 2000,1,10000, 5000)
print('\ntesting2')
testing(2000,1,10000, 2000,1,10000, 12000)
```
