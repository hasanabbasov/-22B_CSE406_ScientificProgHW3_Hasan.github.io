```python
#a). A node with constructor, which takes one argument, value, and initializes the left and right children to None.
class Node:
def __init__(self, key):
self.key = key
self.left = None
self.right = None


#b). Print tree's Inorder traversal.
def inorder(root):
if root is not None:
# Traverse left
inorder(root.left)

# Traverse root
print(str(root.key) + "->", end=' ')

# Traverse right
inorder(root.right)


#c).A function that inserts a new value in the tree at the right location.


def insert(node, key):

# Return a new node if the tree is empty
if node is None:
return Node(key)

# Traverse to the right place and insert the node
if key < node.key:
node.left = insert(node.left, key)
else:
node.right = insert(node.right, key)

return node
  
# d).function that looks up a value in the tree.

def search(node, val):
if val < node.key:
if node.left is None:
return str(val)+" Not Found"
  
node.left=search(node.left,val)
elif val > node.key:
if node.right is None:
return str(val)+" Not Found"
node.right=search(node.right,val)
else:
print(str(node.key) + ' is found\n')

# helper function to be used in delete function which finds the inorder successor.
def minValue(node):
current = node

# Find the leftmost leaf
while(current.left is not None):
current = current.left

return current

# e). function that removes a value from the tree.
def delete(root, key):

# Return if the tree is empty
if root is None:
return root

# Find the node to be deleted
if key < root.key:
root.left = delete(root.left, key)
elif(key > root.key):
root.right = delete(root.right, key)
else:
# If the node is with only one child or no child
if root.left is None:
temp = root.right
root = None
return temp

elif root.right is None:
temp = root.left
root = None
return temp

# If the node has two children,
# place the inorder successor in position of the node to be deleted
temp = minValue(root.right)

root.key = temp.key

# Delete the inorder successor
root.right = delete(root.right, temp.key)

return root
#driver code
root = None
root = insert(root, 11)
root = insert(root, 23)
root = insert(root, 15)
root = insert(root, 16)
root = insert(root, 84)
root = insert(root, 6)
root = insert(root, 1)
root = insert(root, 18)

print("Inorder traversal: ", end=' ')
inorder(root)
#search(root,16)
print("\nDelete 16")
root = delete(root, 16)
print("Inorder traversal: ", end=' ')
inorder(root)
```
